```python
input_mod1 = ad.read_h5ad('../data/Cite_GEX.test.h5ad')
input_mod2 = ad.read_h5ad('../data/Cite_ADT.test.h5ad')
input_mod1, input_mod2 = feature_selection_cite(input_mod1, input_mod2, path='./utils')
```

### Cell matching

```python
out_x, out_y, class_x, class_y = SiaNN.test()
```

```python
merge = np.concatenate((out_x, out_y), axis=0)
merge = pd.DataFrame(merge)
merge.insert(0, 'index', np.concatenate((input_mod1.obs_names.to_numpy(), 
                                         input_mod2.obs_names.to_numpy()), axis=0))
merge.to_csv('SiaNN_embed.txt', index=False)
```

```python
n_neighbors = min(5, len(out_x))
match_matrix = pairwise_distances(out_x, out_y, metric="euclidean")
match_matrix = np.exp(-match_matrix)

batch_code_1 = test_dat.input_mod1.obs['batch'].cat.codes.to_numpy()
batch_code_2 = test_dat.input_mod2.obs['batch'].cat.codes.to_numpy()
# Get the mask of comparisons in a vectorized manner using broadcasting
mask = batch_code_1[:,None] == batch_code_2
# Select the elements other than diagonal ones
mask = mask * 1
match_matrix = match_matrix * mask
#match_matrix = match_matrix[batch_code_1==0,:]
#match_matrix = match_matrix[:,batch_code_1==0]

def top_k_values(array):
    indexes = array.argsort()[-n_neighbors:][::-1]
    A = set(indexes)
    B = set(list(range(array.shape[0])))
    array[list(B.difference(A))]=0
    return array

pairing_matrix = np.apply_along_axis(top_k_values, 1, match_matrix)
# Normalize values to sum to 1
pairing_matrix = pairing_matrix / np.sum(pairing_matrix, axis=1)[:, np.newaxis]
np.trace(pairing_matrix)
#v1: 
#v2: 179 - gene norm
#v3: 160 - gene norm gex, cell norm adt
#v4: 227 - all cell norm
```

### Cell-to-cell matching

```python
match_matrix = pairwise_distances(out_x, out_y, metric="euclidean")
#match_matrix = np.exp(-match_matrix)

batch_code_1 = test_dat.input_mod1.obs['batch'].cat.codes.to_numpy()
batch_code_2 = test_dat.input_mod2.obs['batch'].cat.codes.to_numpy()

match_matrix_1 = match_matrix[batch_code_1==0,:]
match_matrix_1 = match_matrix_1[:,batch_code_2==0]

match_matrix_2 = match_matrix[batch_code_1==1,:]
match_matrix_2 = match_matrix_2[:,batch_code_2==1]
```

```python
neighbor_1 = np.argsort(match_matrix_1, axis=1, kind='stable')
neighbor_1 = np.argsort(neighbor_1, axis=1, kind='stable')

writer = open('../results/raw/neighborhood_s1d2_SiaNN.txt','w')
writer.write('\n'.join(np.diag(neighbor_1).astype('str').tolist()))
writer.close()
```

```python
neighbor_2 = np.argsort(match_matrix_2, axis=1, kind='stable')
neighbor_2 = np.argsort(neighbor_2, axis=1, kind='stable')

writer = open('../results/raw/neighborhood_s3d7_SiaNN.txt','w')
writer.write('\n'.join(np.diag(neighbor_2).astype('str').tolist()))
writer.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100

plt.clf()
fig, ax = plt.subplots(figsize=(3,3))

# batch 1
Z = np.diag(neighbor_1) / neighbor_1.shape[0]
H,X1 = np.histogram( Z, bins = 1000, normed = True )
dx = X1[1] - X1[0]
F1 = np.cumsum(H)*dx

# batch 1
Z = np.diag(neighbor_2) / neighbor_2.shape[0]
H,X2 = np.histogram( Z, bins = 1000, normed = True )
dx = X2[1] - X2[0]
F2 = np.cumsum(H)*dx

plt.plot(X1[1:], F1)
plt.plot(X2[1:], F2)
ax.set_xlabel("Number of neighbors / cells")
ax.set_ylabel("Neighborhood overlap")
ax.legend(['s1d2', 's3d7'])
plt.xlim(-0.05,1)
plt.show()
```

### Label transfer accuracy

```python
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier

def KNN_acc(representations, labels, nn=5):
    indeces = np.arange(representations.shape[0])
    #representations = StandardScaler().fit_transform(representations)
    #pca = PCA(n_components=10)
    #pca.fit(representations[indeces])
    #X_reduced = pca.transform(representations[indeces])
    X_reduced = representations
    neigh = KNeighborsClassifier(n_neighbors=nn, weights='distance')
    n = int(len(representations) // 2)
    neigh.fit(X_reduced[0:n,:], labels)
    ylabel = neigh.predict(X_reduced[n:,:])
    check = len(ylabel) == n
    #print(check)
    acc = np.mean(labels == ylabel)
    print("Accuracy is %.5f"%(acc))
    return ylabel
```

```python
merge = np.concatenate((out_y, out_x), axis=0)
labels = test_dat.input_mod1.obs['cell_type'].cat.codes.to_numpy()
```

```python
#for i in range(5,55,5):
#    print(i)
#    pred_y = KNN_acc(merge, labels, nn=i)
#    print()
#final version
pred_y = KNN_acc(merge, labels, nn=20)
```
